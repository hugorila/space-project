**2. Cleaning the data**

```
# drop columns Unnamed: 0 and Unnamed: 0.1
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
```

```
#Creates country columns
df['Country'] = df.Location.apply(lambda x: x.split(',')[-1].lstrip())
#I called it area, because some places are not states
df['Area'] = df.Location.apply(lambda x: x.split(',')[-2].lstrip())
```

```
#converts Datum into datatype
df['Datum'] = df.Datum.apply(pd.to_datetime, utc=True)

#creates Year, Month and Time Column from Datum
df['Year'] = df.Datum.dt.year
df['Month'] = df.Datum.dt.month
df['Time'] = df.Datum.dt.time
```

```
df['Status Rocket'].replace({'StatusActive':1, 'StatusRetired':0}, inplace=True)
```
